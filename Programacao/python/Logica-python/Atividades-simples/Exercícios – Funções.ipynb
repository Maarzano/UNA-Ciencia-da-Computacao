# Faça um programa contendo funções que recebam:
1. Um número inteiro e que retorne o dobro deste número
2. O ano de nascimento de uma pessoa, e um ano base, e que retorne a idade da pessoa no
ano base
3. Dois números inteiros e que retorne o maior entre eles
4. Um número inteiro e que retorne o valor do fatorial deste número
5. Um número inteiro e que retorne a soma dos números pares entre 1 e o número (inclusive)
6. Um número inteiro e positivo, que retorne a soma dos elementos inteiros existentes entre
1 e o número (inclusive)
7. Dois números inteiros e que retorne a multiplicação do primeiro pelo segundo, utilizando
somente operações de soma.
8. Dois números inteiros e que retorne a potência do primeiro elevado ao segundo, utilizando
somente operações de multiplicação.
9. Um número inteiro e que retorne True se o número for par, e False caso contrário
10. Um número inteiro e que retorne True se o número for primo, e False caso contrário

```python
"""1"""
def dobro(*args):
  re = []
  for i in range(len(args)):
    re.append(args[i] * 2)
  return re
print(dobro(120, 12))
```

```python
"""2"""

def idade(ano_nascimento: int, ano_base: int) -> int:
    return ano_base - ano_nascimento

print(idade(2000, 2025))
```

```python
"""3"""
def maior(*args: float | int):
  m = args[0]
  for valor in args:
    if valor > m:
      m = valor
  return m
```

```python
"""4"""
def fatorial(n: int) -> int:
    if n < 0:
        raise ValueError("Fatorial não definido para números negativos.")
    resultado = 1
    for i in range(2, n + 1):
        resultado *= i
    return resultado

print(fatorial(5))
```

```python
"""5"""
def soma_pares(n: int) -> int:
    return sum(i for i in range(2, n + 1, 2))

print(soma_pares(10))
```

```python
"""6"""
def soma_ate(n: int) -> int:
    if n < 1:
        raise ValueError("O número deve ser positivo.")
    return n * (n + 1) // 2

print(soma_ate(10))
```

```python
"""7"""
def multiplicacao(a: int, b: int) -> int:
    negativo = a < 0 or b < 0
    a, b = abs(a), abs(b)
    resultado = sum(a for _ in range(b))
    return -resultado if (a < 0) ^ (b < 0) else resultado

print(multiplicacao(3, 4))
```

```python
"""8"""
def potencia(base: int, expoente: int) -> int:
    if expoente < 0:
        raise ValueError("Expoente negativo não é suportado neste contexto.")
    resultado = 1
    for _ in range(expoente):
        resultado *= base
    return resultado

print(potencia(2, 5))
```

```python
"""9"""
def eh_par(n: int) -> bool:
    return n % 2 == 0

print(eh_par(4))
print(eh_par(7))
```

```python
"""10"""
def eh_primo(n: int) -> bool:
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

print(eh_primo(7))
print(eh_primo(12))
```
