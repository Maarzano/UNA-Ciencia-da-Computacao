```python
def imprimir_padrao_1(n):
    for i in range(1, n + 1):
        linha = " ".join([str(i)] * i)
        print(linha)

if __name__ == "__main__":
    try:
        n_usuario = int(input("Digite um valor inteiro (n): "))
        if n_usuario < 1:
            print("Por favor, digite um número inteiro positivo.")
        else:
            print("\nResultado (Ex. 1):")
            imprimir_padrao_1(n_usuario)
    except ValueError:
        print("Entrada inválida. Digite um número inteiro.")
```

# 2)

```python
def imprimir_padrao_2(n):
    for i in range(1, n + 1):
        numeros = list(range(1, i + 1))
        linha = " ".join(map(str, numeros))
        print(linha)

if __name__ == "__main__":
    try:
        n_usuario = int(input("\nDigite um valor inteiro (n) para o Ex. 2: "))
        if n_usuario < 1:
            print("Por favor, digite um número inteiro positivo.")
        else:
            print("\nResultado (Ex. 2):")
            imprimir_padrao_2(n_usuario)
    except ValueError:
        print("Entrada inválida. Digite um número inteiro.")
```

```python
def somar_tres(a, b, c):
    return a + b + c

if __name__ == "__main__":
    try:
        num1 = float(input("Digite o primeiro número: "))
        num2 = float(input("Digite o segundo número: "))
        num3 = float(input("Digite o terceiro número: "))

        resultado = somar_tres(num1, num2, num3)
        print(f"A soma de {num1}, {num2} e {num3} é: {resultado}")
    except ValueError:
        print("Entrada inválida. Digite números válidos.")
```

# 4)

```python
def positivo_ou_negativo(argumento):
    if argumento > 0:
        return 'P'
    else:
        return 'N'

if __name__ == "__main__":
    try:
        valor = float(input("Digite um número: "))
        caractere_retornado = positivo_ou_negativo(valor)
        print(f"O caractere retornado para {valor} é: '{caractere_retornado}'")

        print(f"Teste com 10: '{positivo_ou_negativo(10)}'")
        print(f"Teste com 0: '{positivo_ou_negativo(0)}'")
        print(f"Teste com -5: '{positivo_ou_negativo(-5)}'")
    except ValueError:
        print("Entrada inválida. Digite um número válido.")
```

```python
def somaImposto(taxaImposto, custo):
    fator_imposto = 1 + (taxaImposto / 100)
    custo_com_imposto = custo * fator_imposto
    return custo_com_imposto

if __name__ == "__main__":
    print("\n--- Exercício 5: somaImposto ---")
    try:
        taxa = float(input("Digite a taxa de imposto sobre vendas (em %): "))
        custo_inicial = float(input("Digite o custo do item (antes do imposto): "))

        custo_final = somaImposto(taxa, custo_inicial)
        
        print(f"O custo de R$ {custo_inicial:.2f} com uma taxa de imposto de {taxa}% é de R$ {custo_final:.2f}.")
        
        print(f"Custo de R$100.00 com 10% de imposto: R$ {somaImposto(10, 100.00):.2f}")
        print(f"Custo de R$50.00 com 25% de imposto: R$ {somaImposto(25, 50.00):.2f}")
    except ValueError:
        print("Entrada inválida. Digite números válidos.")
```
