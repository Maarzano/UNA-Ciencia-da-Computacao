```python
def contaVogais(texto):
    vogais = 'aeiou'
    contagem_vogais = {}

    for char in texto.lower():
        if char in vogais:
            contagem_vogais[char] = contagem_vogais.get(char, 0) + 1
            
    return contagem_vogais

print("--- Teste da Função contaVogais ---")
texto_entrada = input("Digite um texto para contar as vogais: ")

resultado = contaVogais(texto_entrada)

if not resultado:
    print("Nenhuma vogal foi encontrada no texto.")
else:
    print("\nContagem de vogais:")
    for vogal, quantidade in resultado.items():
        print(f"{vogal}: {quantidade}")
```

```python
def incluirNovoNome(agenda, nome, telefones):
    if nome in agenda:
        print(f"Erro: {nome} já existe na agenda.")
    else:
        agenda[nome] = telefones
        print(f"{nome} foi adicionado(a) com {len(telefones)} telefone(s).")

def incluirTelefone(agenda, nome, telefone):
    if nome in agenda:
        if telefone in agenda[nome]:
            print(f"Erro: O telefone {telefone} já está cadastrado para {nome}.")
        else:
            agenda[nome].append(telefone)
            print(f"Telefone {telefone} adicionado para {nome}.")
    else:
        print(f"{nome} não encontrado(a) na agenda.")
        resposta = input(f"Deseja incluir {nome} na agenda? (s/n): ").lower()
        if resposta == 's':
            incluirNovoNome(agenda, nome, [telefone])

def excluirTelefone(agenda, nome, telefone):
    if nome in agenda:
        if telefone in agenda[nome]:
            if len(agenda[nome]) == 1:
                print(f"Este é o único telefone de {nome}.")
                excluirNome(agenda, nome)
            else:
                agenda[nome].remove(telefone)
                print(f"Telefone {telefone} removido de {nome}.")
        else:
            print(f"Erro: O telefone {telefone} não foi encontrado para {nome}.")
    else:
        print(f"Erro: {nome} não encontrado(a) na agenda.")

def excluirNome(agenda, nome):
    if agenda.pop(nome, None) is not None:
        print(f"{nome} foi removido(a) da agenda.")
    else:
        print(f"Erro: {nome} não encontrado(a) na agenda.")

def consultarTelefone(agenda, nome):
    telefones = agenda.get(nome)
    
    if telefones:
        return telefones
    else:
        print(f"{nome} não encontrado(a) na agenda.")
        return None


def menu_agenda():
    agenda_telefonica = {}
    
    while True:
        print("\n--- Agenda Telefônica ---")
        print("1: Incluir Novo Nome")
        print("2: Incluir Telefone (em nome existente)")
        print("3: Excluir Telefone")
        print("4: Excluir Nome")
        print("5: Consultar Telefones")
        print("6: Ver Agenda Completa")
        print("0: Sair")
        
        opcao = input("Escolha uma opção: ")
        
        if opcao == '1':
            nome = input("Digite o nome: ")
            fones_str = input("Digite o(s) telefone(s) (separados por vírgula): ")
            telefones = [f.strip() for f in fones_str.split(',')]
            incluirNovoNome(agenda_telefonica, nome, telefones)
            
        elif opcao == '2':
            nome = input("Digite o nome: ")
            telefone = input("Digite o novo telefone: ")
            incluirTelefone(agenda_telefonica, nome, telefone)
            
        elif opcao == '3':
            nome = input("Digite o nome: ")
            telefone = input("Digite o telefone a excluir: ")
            excluirTelefone(agenda_telefonica, nome, telefone)
            
        elif opcao == '4':
            nome = input("Digite o nome a excluir: ")
            excluirNome(agenda_telefonica, nome)
            
        elif opcao == '5':
            nome = input("Digite o nome a consultar: ")
            telefones = consultarTelefone(agenda_telefonica, nome)
            if telefones:
                print(f"Telefones de {nome}: {', '.join(telefones)}")
                
        elif opcao == '6':
            print("\n--- Agenda Completa ---")
            if not agenda_telefonica:
                print("Agenda está vazia.")
            else:
                for nome, telefones in agenda_telefonica.items():
                    print(f"- {nome}: {', '.join(telefones)}")
                    
        elif opcao == '0':
            print("Saindo da agenda...")
            break
            
        else:
            print("Opção inválida. Tente novamente.")

print("\n\n--- Teste da Agenda Telefônica ---")
menu_agenda()
```
