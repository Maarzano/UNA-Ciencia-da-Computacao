# FOR

1. Leitura de 5 valores e exibição do dobro

```python
print("=== Dobro de 5 valores ===")
for i in range(5):
    valor = float(input(f"Digite o {i+1}º valor: "))
    print(f"O dobro de {valor} é {valor * 2}")
```

2. Números ímpares até o número informado

```python
print("\n=== Números ímpares até N ===")
n = int(input("Digite um número: "))
print("Números ímpares:")
for i in range(1, n + 1):
    if i % 2 != 0:
        print(i, end=" ")
print()
```

3. Tabuada de um número

```python
print("\n=== Tabuada ===")
num = int(input("Digite um número para ver sua tabuada: "))
for i in range(1, 11):
    print(f"{num} X {i} = {num * i}")
```

4. Números entre dois inteiros

```python
print("\n=== Intervalo entre dois números ===")
a = int(input("Digite o primeiro número: "))
b = int(input("Digite o segundo número: "))

inicio = min(a, b)
fim = max(a, b)

print(f"Números entre {inicio} e {fim}:")
for i in range(inicio, fim + 1):
    print(i, end=" ")
print()
```

5. Soma dos números entre 1 e N

```python
print("\n=== Soma de 1 até N ===")
n = int(input("Digite um número: "))
soma = 0
for i in range(1, n + 1):
    soma += i
print(f"A soma de 1 até {n} é {soma}")
```

6. Fatorial de um número

```python
print("\n=== Fatorial ===")
n = int(input("Digite um número para calcular o fatorial: "))
fatorial = 1
for i in range(1, n + 1):
    fatorial *= i
print(f"{n}! = {fatorial}")
```

7. Média de pesos de 10 pessoas

```python
print("\n=== Média de pesos ===")
soma_pesos = 0
for i in range(10):
    peso = float(input(f"Digite o peso da {i+1}ª pessoa (kg): "))
    soma_pesos += peso
media = soma_pesos / 10
print(f"Média dos pesos: {media:.2f} kg")
```

8. Sexo e peso de 10 pessoas

```python
print("\n=== Análise de peso por sexo ===")
homens = mulheres = 0

for i in range(10):
    sexo = input(f"Sexo da {i+1}ª pessoa (M/F): ").strip().upper()
    peso = float(input("Peso (kg): "))

    if sexo == "M" and 60 <= peso <= 80:
        homens += 1
    elif sexo == "F" and 50 <= peso <= 70:
        mulheres += 1

print(f"Homens com peso entre 60 e 80 kg: {homens}")
print(f"Mulheres com peso entre 50 e 70 kg: {mulheres}")
```

9. Idade e peso de 7 pessoas

```python
print("\n=== Análise de idade e peso ===")
pessoas_90kg = 0
soma_idades = 0
idades = []

for i in range(7):
    idade = int(input(f"Idade da {i+1}ª pessoa: "))
    peso = float(input("Peso (kg): "))

    if peso > 90:
        pessoas_90kg += 1
    soma_idades += idade
    idades.append(idade)

media_idade = soma_idades / 7
print(f"Pessoas com mais de 90 kg: {pessoas_90kg}")
print(f"Média das idades: {media_idade:.2f}")
print(f"Menor idade: {min(idades)}")
print(f"Maior idade: {max(idades)}")
```

10. Verificar se o número é primo

```python
print("\n=== Verificação de número primo ===")
n = int(input("Digite um número: "))
divisores = 0

for i in range(1, n + 1):
    if n % i == 0:
        divisores += 1

if divisores == 2:
    print(f"{n} é um número primo.")
else:
    print(f"{n} não é primo.")
```

11. Multiplicação usando apenas soma

```python
print("\n=== Multiplicação com soma ===")
a = int(input("Digite o primeiro número: "))
b = int(input("Digite o segundo número: "))

resultado = 0
for _ in range(abs(b)):
    resultado += abs(a)

if (a < 0 and b > 0) or (a > 0 and b < 0):
    resultado = -resultado

print(f"{a} x {b} = {resultado}")
```

12. Potência usando apenas multiplicação

```python
print("\n=== Potência com multiplicação ===")
base = int(input("Digite a base: "))
expoente = int(input("Digite o expoente: "))

resultado = 1
for _ in range(expoente):
    resultado *= base
print(f"{base}^{expoente} = {resultado}")
```

# WHILE

1. Validação de ano entre 1900 e 2025

```python
print("\n=== Validação de ano ===")
while True:
    ano = int(input("Informe um ano entre 1900 e 2025: "))
    if 1900 <= ano <= 2025:
        print(f"Ano {ano} é válido.")
        break
    else:
        print("Ano inválido. Tente novamente.")
```

2. Soma de números positivos até digitar um negativo

```python
print("\n=== Soma de números positivos ===")
soma = 0
while True:
    n = float(input("Digite um número positivo (ou negativo para sair): "))
    if n < 0:
        break
    soma += n
print(f"Soma total: {soma}")
```

3. Controle de notas de alunos

```python
print("\n=== Controle de notas ===")
while True:
    matricula = int(input("Digite a matrícula (ou número negativo para sair): "))
    if matricula < 0:
        break

    notas = [float(input(f"Nota {i+1}: ")) for i in range(3)]
    media = sum(notas) / 3

    if media >= 70:
        situacao = "Aprovado"
    elif media >= 60:
        situacao = "Exame"
    else:
        situacao = "Reprovado"

    print(f"Matrícula {matricula} - Média: {media:.2f} - Situação: {situacao}")
```

4. Idades indefinidas com estatísticas

```python
print("\n=== Estatísticas de idades ===")
qtd = soma_idade = acima_50 = abaixo_40 = 0

while True:
    idade = int(input("Digite uma idade (negativa para sair): "))
    if idade < 0:
        break

    qtd += 1
    soma_idade += idade
    if idade > 50:
        acima_50 += 1
    if idade < 40:
        abaixo_40 += 1

if qtd > 0:
    media = soma_idade / qtd
    perc_abaixo_40 = (abaixo_40 / qtd) * 100
    print(f"Pessoas com idade > 50: {acima_50}")
    print(f"Média das idades: {media:.2f}")
    print(f"Percentual com idade < 40: {perc_abaixo_40:.2f}%")
else:
    print("Nenhuma idade informada.")
```

5. Controle de capacidade do elevador

```python
print("\n=== Controle de capacidade do elevador ===")
capacidade = float(input("Informe a capacidade máxima do elevador (kg): "))
peso_total = 0
qtd_pessoas = 0

while peso_total < capacidade:
    peso = float(input("Peso da pessoa (kg): "))
    if peso_total + peso > capacidade:
        print("Capacidade excedida! Pessoa não pode entrar.")
        break
    peso_total += peso
    qtd_pessoas += 1

print(f"Total de pessoas: {qtd_pessoas}")
print(f"Peso total: {peso_total:.2f} kg")
```
